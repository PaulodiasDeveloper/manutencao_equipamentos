```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados = pd.read_excel("Planilha de ocorrências.xlsx")
dados.head(30)
```

```python
# Frequência absoluta
freq_local = dados["Local"].value_counts()
print(freq_local)
```

```python
freq_local
```

```python
# Tabela de frequência (absoluta + relativa)
freq_local_table = pd.DataFrame({
    "Frequência": dados["Local"].value_counts(),
    "Percentual (%)": dados["Local"].value_counts(normalize=True) * 100
})
print(freq_local_table)
```

```python
freq_local_table
```

```python
# Frequência de outra coluna qualitativa
# Por exemplo, Status (Aberto × Fechado):

freq_status = dados["Status"].value_counts()
print(freq_status)
```

```python
# Frequência cruzada (tabela de contingência)
# Para saber quantas ocorrências por Local e Status:

freq_cruzada = pd.crosstab(dados["Local"], dados["Status"])
print(freq_cruzada)
```

```python
# Visualização rápida (opcional)

dados["Local"].value_counts().plot(kind="bar", title="Ocorrências por Local")
plt.xlabel("Local")
plt.ylabel("Frequência")
plt.show()
```
